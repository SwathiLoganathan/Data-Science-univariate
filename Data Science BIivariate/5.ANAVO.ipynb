```python
def QuanQual():
    Quan=[]
    Qual=[]
    for columnName in dataset.columns:
        print(columnName)
        if(dataset[columnName].dtypes=="O"):
            #print("Qual")
            Qual.append(columnName)
        else:
            #print("Quan")
            Quan.append(columnName)
    return Quan,Qual
```

```python
Quan,Qual=QuanQual()
```

# ANAVO : Analysis of Variance

```python
#ONE WAY
import scipy.stats as stats
stats.f_oneway(dataset['ssc_p'],dataset['hsc_p'],dataset['degree_p'])
```

```python
##2. Two Way Classification: Two Independent Variable
import statsmodels.api as sm
from statsmodels.formula.api import ols
anova_two=ols('sl_no ~ ssc_p + hsc_p + degree_p + ssc_p*hsc_p*degree_p',data=dataset).fit()
Two=sm.stats.anova_lm(anova_two,type=2)
Two
```
